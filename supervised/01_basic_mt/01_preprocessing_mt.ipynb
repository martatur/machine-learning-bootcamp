# Importowanie bibliotek

```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

# Generowanie danych

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data)
df_raw
```

# Utworzenie kopii dany

```python
df = df_raw.copy()
df.info()
```

# Zmiana typu danych i wstępne eksploracja

```python
for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()

df.describe()

df.describe().T

df.describe(include=['category']).T
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])

le.fit_transform(df['bought'])
le.classes_

df['bought'] = le.fit_transform(df['bought'])

# powrót do oryginalnej postaci
df['bought'] = le.inverse_transform(df['bought'])
```

# OneHotEndcoder

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output= False)
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
encoder.categories_
```

# get dummies()

```python
df_1 = pd.get_dummies(data=df, drop_first=True, prefix='new')
df_2 = pd.get_dummies(data=df, drop_first=True, prefix_sep='-')
df_3 = pd.get_dummies(data=df, drop_first=True, columns=['size'])
```

# standaryzacja  StandardScaler

```python
df['price']
df['price'].mean()
df['price'].std()

(df['price'] - df['price'].mean()) / df['price'].std()

def standarize(x):
    return (x - x.mean()) / x.std()

standarize(df['price'])
```

```python
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])

scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```
